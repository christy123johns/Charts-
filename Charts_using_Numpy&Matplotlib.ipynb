```python
import numpy as np
import matplotlib.pyplot as plt
Names=np.array(['Anjali','Aswathy', 'Akshaya', 'Mukund', 'Vince'])
Mark=np.array([23,33,32,22,40])
Rank=np.array([4,2,3,5,1])
Gender=np.array(['Female','Female','Female','Male','Male'])
```

```python
plt.bar(Names,Mark)
plt.show()
```

```python
plt.bar(Names, Mark, width = 0.5, edgecolor = 'black', color='lightblue')

plt.title('Marks of Students\n', fontdict= {'fontsize':20, 'fontweight':5, 'color':'Green'})
plt.xlabel('Names', fontdict= {'fontsize':20, 'fontweight':5, 'color':'brown'})
plt.ylabel('Mark', fontdict= {'fontsize':20, 'fontweight':5, 'color':'brown'})
plt.show()
```

```python
plt.scatter(Mark,Rank)
plt.title('Mark and ranks of studnets\n',fontdict={'fontsize':30,'fontweight':0.5,'color':'blue'})
plt.xlabel('mark',fontdict={'fontsize':20,'color':'brown'})
plt.ylabel('rank',fontdict={'fontsize':20,'color':'brown'})
plt.show()
```

```python
plt.scatter(Mark[Gender=='Female'],Rank[Gender=='Female'],alpha=0.7, s=50,c='orange')
plt.title("Mark and rank of females\n",fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'black'})
plt.xlabel("\n Mark w.r.t females", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel("rank w.r.t females", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
plt.show()
```

```python
plt.scatter(Mark[Gender=='Male'],Rank[Gender=='Male'],alpha=0.7, s=50,c='orange')
plt.title("Mark and rank of males\n",fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'black'})
plt.xlabel("\n Mark w.r.t males", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel("rank w.r.t males", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
plt.show()
```
